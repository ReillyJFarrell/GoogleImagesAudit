# 1. Executive Summary

Our study explores the representation of different nationalities, with filtering potential through gender, by looking at the headlines of images pulled from Google Image search. Our study finds a large amount of interesting trends and figures, but due to the lack of a comparable baseline the identification of proper misrepresentation, or over-representation of a particular sentiment, cannot be objectively proven. On the other hand there are interesting conclusions that may be inferred, such as the high frequency of dating headlines regarding woman compared to men (section 10.6), the high density of negatively connotated headlines in Russia/Ukraine and the Middle East (sections 6.1 and 10.4), and a few other speculations that are apparent in the paper. While the conclusions pulled cannot be concrete because of limitations discussed in section 7, there is a good possibility of extensions that could yield interesting results.

**2.1. The Definition of the Harm and the Investigated Service**
The potential harms, biases, etc. that we wanted to look at in this study were the potential biasing and/or misrepresentation of people with demographic filters in regard to nationality and gender. For example were some nationalities sourcing headlines with overtly negative sentiment without relevant reason (for example Russian men had a generally negative sentiment but there's contemporary evidence for why, what about British men?)? Were articles and headlines for sourced images for women of some nationality overtly sexualized or oddly tuned to dating advice? These were the sort of questions we went into the data with.

The main thing to keep in mind with regard to this harm is that there is no definitive baseline for perception like there was with the Google Images job representation study we saw in class. In that study there could be a direct comparison between the BLS data i.e., "the ground truth" and the representation found in the image results. Here we can pull the sentiment data from the headlines, but we have to take the data into a general context, rather than comparison. An interesting extension that we'll talk about later is developing some baseline of comparison via surveys and perceptions of nationalities, although there's a whole slew of ethical considerations there.

The service being investigated is Google Images/the Google Search function since a lot of the image headlines are going to be pulled with some overlap from general search.

**5.2. What are the limitations of the data and what kind of additional information would have helped you to investigate the harm better (even if the missing data is impossible to get)**

The primary limitation our dataset is that we do not have a baseline sentiment dataset to compare our retrieved dataset to. What would be useful to collect would be a study of n persons who give a NEU, POS, or NEG sentiment to different nationalities, genders, the combination, and possibly other keywords such as LGTBQ, etc. By establishing this baseline we could then compare our results to this. For example, the query "british person" had a NEG:POS:NEU split of around 25%:5%:70%. Meanwhile in a map produced by the Dorfius and Swindle study found that the sentiment of Britain (as a country so the baseline doesn't match perfectly which is why we don't use it as our actual baseline/comparator) was pretty positive:

**6.3. An aside on keyword specification**

In order to slice the data by keywords we used Regex to find tokens in the pulled headlines such that we could count the instances. For some filters the keywords were self-explanatory. For example in searching for text associated with "ugly" we used the following:

```
=ISNUMBER(SEARCH("ugly",[@Text]))
```

which just looks for the keyword ugly.

Other filters had a little more ambiguity. One such example is the LGTBQ filter we came up with, which we decided to just have incorporate the different words in the acronym:

```
=IF(regextest([@Text],"\b(lgbtq?|gay|lesbian|bisexual|queer|transgender)\b"),TRUE,FALSE)
```

The final level of ambiguity to overcome were with filters like "sex". Since we wanted to get a decent idea of the headlines that were associated with this, we couldn't just use the word "sex" because this would be too narrow of a scope. We used associated words to include in the regex to get a broader picture:

```
=IF(regextest([@Text],"\b(sex|sexy|sexiest|sexier|sexiness|sexual|sexually|hot|hottest|hotties?)\b"),TRUE,FALSE)
```

Now the important point: A proper extension of this study should include a more objective way of coming up with associated search terms. Because of us coming up with associated words, there may be some implicated bias. Of course, we tried to remain as objective as possible but you can never really be certain unless you use methods that provide this proof of certainty.

**10.5. Net Sentiment**

Here we're looking at net sentiment, or the difference between the negative and positive sentiment. To do this we groupby the sentiments from queries, and then aggregate again to get rid of the gendered data. This allows us to get the net sentiment for each country.

````python
sentiments_by_query = data.groupby(["iso_a3", "last_word"])["Sentiment"].value_counts() / data.groupby(["iso_a3", "last_word"])["Sentiment"].count()
````

````python
net_sentiments_by_query = data.groupby(["iso_a3", "last_word"])["Sentiment"].aggregate(lambda x: (sum(1 for a in x if a == "POS") - sum(1 for a in x if a == "NEG"))/ sum(1 for a in x if a == "POS" or a == "NEG") if sum(1 for a in x if a == "POS" or a == "NEG") != 0 else 1)
````

````python
net_sentiments_by_query["FRA"]["man"]
````

````python
net_sentiments_by_country = data.groupby("iso_a3")["Sentiment"].aggregate(lambda x: (sum(1 for a in x if a == "POS") - sum(1 for a in x if a == "NEG"))/ sum(1 for a in x if a == "POS" or a == "NEG") if sum(1 for a in x if a == "POS" or a == "NEG") != 0 else 1)
````

*Explorable Net Sentiment by Country Graph*

````python
world["net_sentiment"] = world["iso_a3"].map(net_sentiments_by_country)
world.explore(column="net_sentiment", legend=True)
````

Here we see something similar to the negativity map, where there seems to be a lower sentiment in the middle east, Russia, and central Africa.

**10.6 Dating add-ons**

In the keyword filtering section we talked about some limitations in terms of getting keyword frequency by gender. Well now we fixed that! We used regex again and filtered by keywords, then were able to graph by gender.

````python
import re
dating_re = re.compile(r"\b(date|dating)\b", re.I)

data["dating"] = data["Text"].map(lambda t: 1 if isinstance(t, str) and dating_re.search(t) else 0)
````

````python
dating_by_gender = data.groupby("last_word")["dating"].sum()
dating_by_gender
````

````python
plt.bar(["man", "person", "woman"], dating_by_gender)
plt.title('Dating Keyword Frequency by Gender')
plt.xlabel('Query')
plt.ylabel('Frequency of Keyword')
plt.show()
````

Here we can see that the hypothesis we had originally is true. There is a higher frequency of dating related headlines for woman than there are for men.

````python
dating_by_country = data.groupby("iso_a3")["dating"].sum() / data.groupby("iso_a3")["dating"].count()
````

````python
world["dating_pct"] = world["iso_a3"].map(dating_by_country)
````

*Explorable Dating Distribution Map*

````python
world.explore(column="dating_pct", legend=True)
````

Now here's where it gets interesting. Besides the obvious outlier in Argentina, there seems to be a trend of dating keyword density being higher in central to Eastern Europe, as well as countries in South America commonly stereotyped with attractiveness (although Brazil oddly enough isn't as high as one would think).
